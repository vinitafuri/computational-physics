# Derivadas

Método numérico de cálculo de derivadas. A repetição é dada pela relação de recorrência:

$$ f'(x_n) = \frac{f(x_{n+1}) - f(x_{n-1}) }{2h}$$

Com $h$ sendo o intervalo entre os incrementos. Existem diversas outras formas de se calcular a derivada numericamente.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def derivada(h,x):
    return (f(x+h)-f(x-h))/(2*h)

N=50
pontos=np.linspace(0,4*np.pi,N)
fx=f(pontos)
h=pontos[1]-pontos[0]
dfx=derivada(h,pontos)

deriv=np.zeros(N)
mseno=np.zeros(N)
for i in range(1,N-1):
    deriv[i]=(fx[i+1]-fx[i-1])/(2*h)
    mseno[i]=-np.sin(h*i)

plt.plot(pontos,dfx,label=r'$sen(x)$ numérico')
plt.plot(pontos,mseno,label=r'$sen(x)$')
plt.plot(pontos, f(pontos),label=r'$cos(x)$')
plt.xlabel(r'x')
plt.ylabel(r'f(x)')
plt.legend(loc='best')
```
